```python
import os
import glob
import pandas as pd
import numpy as np
```

```python
data_input_dir = r'C:\projects\CONUS_TGW_WRF_SSP585_HOT\CONUS_TGW_WRF_SSP585_HOT'

ba_code = 'AEC'

# Create a vector of temperature bins used to compute the histogram:
t_bins = np.arange(250,350,5)

t_bins
```

```python
# Make a list of the files associated with a single BA indicated by "ba_code":
list_of_files = glob.glob(os.path.join(data_input_dir, f'{ba_code}_*.csv'))

list_of_files
```

```python
# Loop over the range of files in the filelist:
for i in range(len(list_of_files)):
    # Read in the data for a single year:
    df = pd.read_csv(list_of_files[i], index_col = None, header = 0)
    
    # Compute the histogram using the t_bins vector to define the bin edges:
    hist, bin_edges = np.histogram(df['T2'], bins=t_bins, density=True)
    
    # Convert the numpy array to a df:
    hist_df = pd.DataFrame(hist, columns=[str(i)])
    
    # Aggregate the output into a new dataframe:
    if i == 0:
        aggregate_output_df = hist_df
    else:
        aggregate_output_df = pd.concat([aggregate_output_df, hist_df], axis=1)
    
# Display the aggregate df with each column being the frequency of T2 for a single year of data:
aggregate_output_df
```
